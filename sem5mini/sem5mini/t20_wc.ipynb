```python
import pandas as pd
import json
```

```python
with open('t20_json_files/t20_wc_match_results.json')as f:
    data=json.load(f)
df_match=pd.DataFrame(data[0]['matchSummary'])

df_match.head(11)
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace = True)
df_match.head()
```

```python
df_match.to_csv('t20_csv_files/match_results.csv',index=False)
```

```python
match_ids_dicts={}
for index,row in df_match.iterrows():
    key1=row['team1'] + ' Vs ' +row['team2']
    key2=row['team2'] + ' Vs ' +row['team1']
    match_ids_dicts[key1]=row["match_id"]
    match_ids_dicts[key2]=row["match_id"]
```

```python
with open('t20_json_files/t20_wc_batting_summary.json')as d:
    data=json.load(d)
    all_record=[]
for rec in data:
    all_record.extend(rec['battingSummary'])
df_batting=pd.DataFrame(all_record)
df_batting.tail()
        
```

```python
df_batting['out/not_out']=df_batting.dismissal.apply(lambda x:"out" if len(x)>0 else "not_out")
df_batting.head(11)
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
df_batting.head(11)
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x:x.replace('â€',''))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x:x.replace('\xa0',''))
df_batting.head(11)
```

```python
df_batting["match_id"]=df_batting['match'].map(match_ids_dicts)
df_batting.head(11)
```

```python
df_batting.to_csv('t20_csv_files/fact_batting_summary.csv',index=False)
```

```python
with open('t20_json_files/t20_wc_bowling_summary.json')as d:
    data=json.load(d)
    all_record=[]
for rec in data:
    all_record.extend(rec['bowlingSummary'])
df_bowling=pd.DataFrame(all_record)
df_bowling.head(5)
```

```python
df_bowling["match_id"]=df_bowling['match'].map(match_ids_dicts)
df_bowling.head(11)
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv',index=False)
```
